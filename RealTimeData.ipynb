```python
%matplotlib notebook
import time
from math import sin
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()

i = 0

period = 1.0 / 1000
i = 0
timestamps = []
raw_sigs = []
time_start = time.time()

while True:
    now = time.time()
    #raw_sig = sin(2 * np.pi * 3 * i) + sin(2 * np.pi * 0.5 * i)
    #raw_sig = sin(now - time_start)
    raw_sig = 2 * sin(2 * np.pi * 1000 * (now - time_start))
    timestamps.append(now - time_start)
    raw_sigs.append(raw_sig)
    i += period
        
    ax.plot(timestamps, raw_sigs, color='b')
    fig.canvas.draw()
    time.sleep(0.1)
```
